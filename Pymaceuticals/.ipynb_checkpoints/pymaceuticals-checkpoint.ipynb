```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
combined_df

# Display the data table for preview
study_df =combined_df[["Mouse ID","Timepoint", "Tumor Volume (mm3)","Metastatic Sites","Drug Regimen", "Sex", "Age_months","Weight (g)"]]
study_df.head()
```

```python
# Checking the number of mice.
mice_count=mouse_metadata["Mouse ID"].count()
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice =study_df[study_df.duplicated(['Mouse ID', 'Timepoint'], keep=False)]
duplicate_ids = duplicate_mice['Mouse ID'].unique()
duplicate_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data = study_df.loc[study_df['Mouse ID'] == 'g989']
duplicate_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = study_df.drop(study_df.loc[study_df['Mouse ID'] == 'g989'].index)
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
count_after_del = len(clean_df['Mouse ID'].unique())
count_after_del
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
main_df = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"]
mean= main_df.mean()
median=main_df.median()
standard_dev = main_df.std()
variance = main_df.var()
standard_err = main_df.sem()

summary_stats = pd.DataFrame({  "Mean Tumor Volume": mean,
                                "Median Tumor Volume": median,
                                "Tumor Volume Variance": variance,
                                "Tumor Volume Std. Dev.": standard_dev,
                                "Tumor Volume Std. Err": standard_err})
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Group the data by "Drug Regimen" and calculate the summary statistics in a single line using .agg()
agg_df = clean_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean","median","var","std","sem"]})
agg_df
```

## Bar and Pie Charts

```python
# Grouping the data by 'Drug Regimen' and counting the number of rows
regimen_counts = clean_df.groupby('Drug Regimen').size()
regimen_counts = regimen_counts.sort_values(ascending=False)

# Creating the bar plot using pandas
regimen_counts.plot(kind='bar', color="blue")

# Setting the labels and title
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Time Points")


# Displaying the plot
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.

counts = clean_df.groupby("Drug Regimen").size()
counts = counts.sort_values(ascending=False)
# Create a bar plot
plt.bar(counts.index, counts.values, color ='b')

# Set the x-axis label
plt.xlabel("Drug Regimen")
plt.xticks(rotation="vertical")
# Set the y-axis label
plt.ylabel("# of Observed Mouse Time Points")
plt.show()
```

```python
gender_counts=clean_df["Sex"].value_counts()
gender_counts.plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.ylabel('Sex')
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_df.groupby("Mouse ID")["Timepoint"].max()
last_timepoint
merged_df = pd.merge(last_timepoint, clean_df, on=["Mouse ID", "Timepoint"], how="left")
merged_df
treatment_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
final_tumor_volume = []
for treatment in treatment_regimens:
    treatment_match = merged_df[merged_df["Drug Regimen"] == treatment]
    #print(treatment_match)
    final_tumor_volume.append(treatment_match["Tumor Volume (mm3)"])
print(final_tumor_volume)
```
